```python
import csv
import statistics
from matplotlib import pyplot as plt
```

```python
def lee_fichero(fichero):
    with open (fichero, encoding="utf-8") as f:
        lector = csv.reader(f)
        next(lector)
        accidentes = []
        for index, date, time, location, operator,typee,fatalities, summary in lector:
            index = int(index)
            fatalities = float(fatalities)
            tupla = (index, date, time, location, operator, typee, fatalities, summary)
            accidentes.append(tupla)
    return accidentes

            
#lee_fichero('./data/Fichero.csv')
```

```python
ACCIDENTES = lee_fichero('./data/Fichero.csv')
print(ACCIDENTES[:7])
```

```python
def  mostra_que_accidentes_tienen_registro_de_la_hora(ACCIDENTES):  
    for x in ACCIDENTES:
        if(x[dict["time"]] == "Desconocida"):
            print(x[dict["index"]],"No se sabe.")
        else:
            print(x[dict["index"]],"Hay un registro de la misma.")
```

```python
#Filtrar por estación
```

```python
def clasificar_dias_segun_estacion(ACCIDENTES):
    for x in ACCIDENTES:
        if(x[dict["date"]][1:2]== 6,7,8):
            print(x[dict["date"]],"Verano")
        if(x[dict["date"]][1:2]== 3,4,5):
            print(x[dict["date"]],"Primavera")
        if(x[dict["date"]][0:2]== 9,10,11):
            print(x[dict["date"]],"Otoño")
        else:
            print(x[dict["date"]],"Invierno")
```

```python
clasificar_dias_segun_estacion(ACCIDENTES)
```

```python
def mostrar_los_accidentes_graves(ACCIDENTES):
    for x in ACCIDENTES:
        if(x[dict["fatalities"]]>12):
            print(x[dict["index"]],":","Accidente grave.")
```

```python
def media_de_fallecidos_en_Octubre(ACCIDENTES):
    suma=0
    cont=0
    for x in ACCIDENTES:
        if(x[dict["date"]][0:2]=="10"):
            suma= suma + x[dict["fatalities"]]
            cont = cont + 1
    print(suma)
    print(cont)
    print(suma/cont)
```
