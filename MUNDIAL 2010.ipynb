```python
import csv
from matplotlib import pyplot as plt
from collections import namedtuple, defaultdict
```

```python
#FUNCION DE LECTURA
```

```python
Seguro = namedtuple('Seguro', 'pais, numero_camiseta, posicion nombre_completo apellido nombre nombre_camiseta nacimiento club pais_club altura')
def lee_registros(fichero):
    with open(fichero, 'r', encoding='utf-8') as f:
        lector = csv.reader(f)
        next(lector)
        registros = [Seguro(str(pais), int(numero_camiseta), str(posicion), str(nombre_completo), str(apellido), str(nombre), str(nombre_camiseta), str(nacimiento), str(club), str(pais_club), int(altura))
                    for pais, numero_camiseta, posicion, nombre_completo, apellido, nombre, nombre_camiseta, nacimiento, club, pais_club, altura in lector]
    return registros

            
lee_registros('./squads.csv')
```

```python
REGISTROS = lee_registros('./squads.csv')
print(len(REGISTROS), REGISTROS[:5])
```

```python
#CONSULTA Y FILTRADO
```

```python
def calcula_paises(registros):
    ''' Lista ordenada de regiones distintas
    
    ENTRADA:
       - lista de registros 
    SALIDA:
       - lista de las regiones ordenada alfabéticamente -> [str]
    '''

    return sorted(list({r.pais for r in registros}))
```

```python
print(calcula_paises(REGISTROS))
```

```python
def filtra_por_paises(registros, pais_deseado):
    for registro in registros:
        if registro.pais == pais_deseado:
            print(f'Nombre completo: {registro.nombre_completo}, nº: {registro.numero_camiseta} Posición: {registro.posicion}, ')
```

```python
filtra_por_paises(REGISTROS, 'Italy')
print('')
filtra_por_paises(REGISTROS,'Spain')
print('')
filtra_por_paises(REGISTROS,'Argentina')
```

```python
def obtener_numero_clubes_unicos_por_pais(registros, pais_deseado):
    clubes_unicos = set()
    
    for registro in registros:
        if registro.pais_club == pais_deseado:
            clubes_unicos.add(registro.club)
    
    numero_clubes = len(clubes_unicos)
    return f'Número de clubes en {pais_deseado}: {numero_clubes}'
```

```python
obtener_numero_clubes_unicos_por_pais(REGISTROS, ' NED')
```

```python
obtener_numero_clubes_unicos_por_pais(REGISTROS, ' ESP')
```

```python
obtener_numero_clubes_unicos_por_pais(REGISTROS, ' ENG')
```

```python
def obtener_info_jugador_mas_alto_por_pais(registros, pais_deseado):
    jugador_mas_alto = None
    
    for registro in registros:
        if registro.pais == pais_deseado:
            if jugador_mas_alto is None or registro.altura > jugador_mas_alto.altura:
                jugador_mas_alto = registro
    
    if jugador_mas_alto:
        print(f'Jugador más alto de {pais_deseado}:')
        print(f'Nombre completo: {jugador_mas_alto.nombre_completo}')
        print(f'Altura: {jugador_mas_alto.altura} cm')
        print(f'Número de camiseta: {jugador_mas_alto.numero_camiseta}')
        print(f'Posición: {jugador_mas_alto.posicion}')
        print(f'Club: {jugador_mas_alto.club}')
    else:
        print(f'No se encontraron jugadores de {pais_deseado} en los registros.')
```

```python
obtener_info_jugador_mas_alto_por_pais(REGISTROS, 'Portugal')
```
